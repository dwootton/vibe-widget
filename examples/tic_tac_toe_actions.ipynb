# Tic-Tac-Toe (Actions + Outputs)

This example loads a prebuilt widget and exercises the action interface.
Click any square to place X, then run the next cell to trigger an AI move and read outputs.


```python
import time
from pathlib import Path

import vibe_widget as vw

js_code = Path("examples/tic_tac_toe_widget.js").read_text()

widget = vw.create(
    "Tic-tac-toe with actions",
    outputs=vw.outputs(
        board_state="current board state",
        game_over="game over flag",
        current_turn="current player turn",
    ),
    actions=vw.actions(
        ai_move=vw.action("AI move", params={"index": "0-8"})
    ),
    display=True,
    cache=False,
)

widget.code = js_code
widget
```

## Simulate a turn change

After you click a square (X move), run this cell to dispatch an AI move and verify outputs.


```python
widget.actions.ai_move(index=4)
time.sleep(0.2)
widget.board_state, widget.current_turn, widget.game_over
```
